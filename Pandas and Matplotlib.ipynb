```python
import numpy as np
import matplotlib.pyplot as plt

overs=np.arange(5,50,5)
overs1=np.arange(5,30,5)
runs=np.array([25,51,84,131,86,189,220,250,267])
runs1=np.array([15,41,94,110,151])
wk=np.array([12,32,96])
plt.plot(overs,runs,color="blue",label="india")
plt.plot(overs1,runs1,color="Yellow",label="Aus")
plt.legend(loc='best')
plt.show()
```

```python
a=[230,560,780,127,128]
b=[200,160,270,127,400]
years=[1,2,3,4]
profit_a=[(a[i]-a[i-1]) for i in range(1,len(a))]
profit_b=[(b[i]-b[i-1]) for i in range(1,len(b))]
plt.subplot(2,1,2)
plt.plot(years,profit_a,color='hotpink',linewidth='3',label='CompanyA',marker='*',ms='15',mec='k')
plt.subplot(2,1,1)
plt.plot(years,profit_b,color='black',linestyle='dotted',label='CompanyB',marker='h')
plt.legend(loc='best')
plt.show()
```

```python
a=np.array([25,60,5,10])
label=["ARRAY","PYTHON","PANDAS","NUMPY"]
explo=[0,0,0.3,0.4]
color=["red","orange","pink","green"]
plt.pie(a,labels=label,explode=explo,startangle=180,colors=color,autopct="%1.2f%%",textprops={"fontsize":13,"color":'blue'})
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd


df=pd.read_excel("/content/temp.xlsx")
plt.plot(df['day'], df['temperature'],marker='o')
temp=np.arange(15,df['temperature'].max()+2,5)
temp1=np.arange(1,df['day'].max()+2,2)
plt.xlabel('Days')
plt.ylabel('Temperature (°C)')


plt.title('Temperature analysis')
threshold = 35.0
days_above_threshold = df[df['temperature'] > threshold]['day'].count()
print(f'Number of days where temperature exceeded {threshold}°C: {days_above_threshold}')
print('average temperature is',df['temperature'].mean())
print("Mininum temperature is",df['temperature'].min())
print("Maximum temperature is",df['temperature'].max())
plt.show()
```

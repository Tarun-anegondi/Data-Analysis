```python
url1='https://www.forbesindia.com/article/explainers/top-10-richest-people-india/85909/1'
page=r.get(url1)
```

```python
page=bs(page.text,"html")
print(page)
```

```python
m=page.find('title')
print(m)
```

```python
print(m.text)
```

```python
v=page.find('table')
print(v)
```

```python
p=v.find_all('th')
print(p)
```

```python
h=[]
for i in p:
  h.append(i.text)
print(h)
```

```python
h=[i.text for i in p]    #list comprehension[lc]
print(h)              #listname=[o/p and forloop]
```

```python
tarun=page.find('table')
```

```python
t=tarun.find_all('th')
print(t)
h=[i.text for i in t]
print(t)
```

```python
# creating dataframe
import pandas as pd
df=pd.DataFrame(h[0:4:])
print(df)
```

```python
#creating data frame
import pandas as pd
df=pd.DataFrame(columns=h)
print(df)
```

```python
rows=v.find_all('tr')
print(rows)
```

```python
ind=0
for i in rows[1:6]:
  #print(i)
  erd=i.find_all('td')
  #print(erd)
  rd=[j.text for j in erd]
  #print(rd)
  df.loc[ind]=rd
  ind=ind+1
```
